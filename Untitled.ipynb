```python
##create sea surface temp figures

import requests
from netCDF4 import Dataset
import xarray as xr

def get_data(url):
    link = Dataset(url)
    return(xr.open_dataset(xr.backends.NetCDF4DataStore(link)))

# grab dataset from 1972 and 1986
sst_1972=get_data('https://www.ncei.noaa.gov/pub/data/cmb/ersst/v5/netcdf/ersst.v5.197207.nc#mode=bytes')
sst_1986=get_data('https://www.ncei.noaa.gov/pub/data/cmb/ersst/v5/netcdf/ersst.v5.198607.nc#mode=bytes')
```

```python
display(sst_1972)
display(sst_1986)
```

```python
import numpy as np
dimentions = sst_1972.dims
#grab lat and long data

sst_1972.lon.values

lat=sst_1972.lat.values
lon=sst_1972.lon.values
print(lat, lon)

display(sst_1972.isel(lev=0, time=0))

ssta_1972=sst_1972['ssta'].isel(lev=0, time=0)
ssta_1986=sst_1986['ssta'].isel(lev=0, time=0)
np.shape(ssta_1972)
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

#create figure object
fig = plt.figure()
ax = plt.axes(projection = ccrs.PlateCarree())

#add coasrline
ax.coastlines(resolution = '110m', color ='navy')

#make heat/color map
plt.pcolormesh(lon, lat, ssta_1972, cmap='magma', transform=ccrs.PlateCarree())

#add colorbar
plt.colorbar()

#title
plt.title('1972 Sea Surface Temp')



#create figure object
fig = plt.figure()
ax = plt.axes(projection = ccrs.PlateCarree())

#add coasrline
ax.coastlines(resolution = '110m', color ='navy')

#make heat/color map
plt.pcolormesh(lon, lat, ssta_1986, cmap='magma', transform=ccrs.PlateCarree())

#add colorbar
plt.colorbar()

#title
plt.title('1986 Sea Surface Temp')
```

```python
##refine graph and make it better

import matplotlib.pyplot as plt
import cmocean
from mpl_toolkits.axes_grid1 import make_axes_locatable
import cartopy.crs as ccrs
import cartopy.feature as cfeature


# Figure for 1972
fig = plt.figure()
ax = plt.axes(projection=ccrs.Robinson())
p = plt.pcolormesh(lon, lat, ssta_1972, cmap=cmocean.cm.deep,
                   transform=ccrs.PlateCarree(), vmin=-4, vmax=4.5)

# Change land color to green
ax.add_feature(cfeature.LAND, color='honeydew')

ax.coastlines(resolution='110m', color='black')

cb = fig.colorbar(p, label=r"Sea Surface Temp (ÂºC)", shrink=0.65, pad=0.05)
plt.title('1972 Sea Surface Temperature')

g1 = ax.gridlines(crs=ccrs.PlateCarree(), color='black', linestyle='--', alpha=0.5)
g1.bottom_labels = True
g1.left_labels = True

# Figure for 1986
fig = plt.figure()
ax = plt.axes(projection=ccrs.Robinson())
p = plt.pcolormesh(lon, lat, ssta_1986, cmap=cmocean.cm.deep,
                   transform=ccrs.PlateCarree(), vmin=-4, vmax=4.5, label='Sea Surface Temp (ÂºC)')

# Change land color to lightgreen
ax.add_feature(cfeature.LAND, color='honeydew')

ax.coastlines(resolution='110m', color='black')

cb = fig.colorbar(p, label=r"Sea Surface Temp (ÂºC)", shrink=0.65, pad=0.05)
plt.title('1986 Sea Surface Temperature')

g2 = ax.gridlines(crs=ccrs.PlateCarree(), color='black', linestyle='--', alpha=0.5)
g2.bottom_labels = True
g2.left_labels = True

plt.show()
```
